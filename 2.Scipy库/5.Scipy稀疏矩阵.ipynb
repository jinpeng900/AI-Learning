# SciPy 稀疏矩阵
稀疏矩阵（英语：sparse matrix）指的是在数值分析中绝大多数数值为零的矩阵。反之，如果大部分元素都非零，则这个矩阵是稠密的(Dense)。

在科学与工程领域中求解线性模型时经常出现大型的稀疏矩阵。

##### 我们主要使用以下两种稀疏矩阵格式：
- CSC - 压缩稀疏列格式（Compressed Sparse Column format）
- CSR - 压缩稀疏行格式（Compressed Sparse Row format）

### CSR 矩阵
- 我们可以通过 scipy.sparse.csr_matrix() 函数传递数组来创建 CSR 矩阵。

```python
# 创建 CSR 矩阵
import numpy as np
from scipy.sparse import csr_matrix

arr = np.array([[0, 1, 0], [0, 0, 1], [1, 0, 2]])

print(csr_matrix(arr))

"""
第 0 行 第 1 列 为非零元素 1
第 1 行 第 2 列 为非零元素 1
第 2 行 第 0 列 为非零元素 1
第 2 行 第 2 列 为非零元素 2
"""
```
