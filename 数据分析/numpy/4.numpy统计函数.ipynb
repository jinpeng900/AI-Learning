### 统计函数
- np.mean() , np.sum() : 所有元素的平均值 , 所有元素的和 , 参数是 number 或 array
- np.max() , np.min() : 所有元素的最大值 , 所有元素的最小值 , 参数是 number 或 array
- np.std() , np.var() : 所有元素的标准差 , 所有元素的方差 , 参数是 number 或 array
- np.argmax() , np.argmin() : 所有元素的最大值的索引 , 所有元素的最小值的索引 , 参数是 number 或 array
- np.cumsum() , np.cumprod() : 所有元素的累计和 , 所有元素的累计积 , 参数是 number 或 array


##### 1.统计函数-所有元素的平均值 , 所有元素的和 , 参数是 number 或 array

```python
import numpy as np
arr = np.arange(15).reshape(3 , 5)
avg = np.mean(arr)
print("数组arr的平均值为:" , avg)
sum_num = np.sum(arr)
print("数组arr的元素和为:" , sum_num)
```

##### 2.统计函数-所有元素的最大值 , 所有元素的最小值 , 参数是 number 或 array

```python
max_num = np.max(arr)
print(f"数组arr的最大值为:{max_num}")
min_num = np.min(arr)
print(f"数组arr的最小值为:{min_num}")
```

##### 3.统计函数-所有元素的标准差 , 所有元素的方差 , 参数是 number 或 array

```python
std_num = np.std(arr)
print(f"数组arr的标准差为:{std_num}")
var_num = np.var(arr)
print(f"数组arr的方差为:{var_num}")
```

##### 4.统计函数-所有元素的最大值的索引 , 所有元素的最小值的索引 , 参数是 number 或 array

```python
max_arg = np.argmax(arr)
print(f"数组arr的最大值的索引为:{max_arg}")
min_arg = np.argmin(arr)
print(f"数组arr的最小值的索引为:{min_arg}")
```

##### 5.统计函数-所有元素的 累计和 或 累计积 , 参数是 number 或 array

```python
cum_sum = np.cumsum(arr)
print(f"数组arr的累计和为:{cum_sum}")
cum_prod = np.cumprod(arr)
print(f"数组arr的累计积为:{cum_prod}")
```
