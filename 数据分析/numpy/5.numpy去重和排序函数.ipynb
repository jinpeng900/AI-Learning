##### 1.去重函数-unique() : 找到数组中的唯一值并返回排序结果 , 类似于python的set()函数

```python
import numpy as np
arr = np.array([[1 , 2 , 1] , [2 , 3 , 4]])
print(f"数组arr的唯一值为:{np.unique(arr)}")
arr1 = np.arange(15).reshape(3 , 5)
print(f"数组arr1的唯一值为:{np.unique(arr1)}")
```

##### 2.排序函数-sort() : 对数组进行排序 , 默认是按行排序 , 也可以按列排序

```python
arr2 = np.array([[10 , 2 , 1] , [8 , 3 , 4]])
arr2.sort()
arr2
```
