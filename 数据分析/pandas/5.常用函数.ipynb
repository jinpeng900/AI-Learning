# 实例

```python
import pandas as pd

# 读取 JSON 文件
df = pd.read_json('data.json')

print(df)
```

```python
# 填补缺失值
x = df['age'].mean()
y = df['score'].mean()
df = df.fillna({'age':x, 'score':y})
print(df)
```

```python
# 重命名列
df = df.rename(columns={'name':'姓名' , 'age':'年龄' , 'gender':'性别' , 'score':'成绩'})
print(df)
```

```python
# 按成绩排序
df = df.sort_values(by = ['成绩' , '年龄'] , ascending=[False,True])
print(df)
```

```python
# 按性别分组并计算平均成绩和年龄
grouped = df.groupby('性别').agg({'成绩':'mean' , '年龄':'mean'})                   # agg 方法可以对多个列进行聚合操作
print(grouped)
```

```python
# 按性别分组并计算平均成绩
grouped  = df.groupby('性别')['成绩'].mean()
print(grouped)
```

```python
# 选择成绩大于等于80的学生 , 并只保留姓名和成绩列
df = df.loc[df['成绩'] >= 80 , ['姓名' , '成绩']]
print(df)
```

```python
# 计算每列的基本统计信息
stats = df.describe()
print(stats)
```

```python
# 计算每列的平均值
df = pd.read_json('data.json')
mean = df[['age' , 'score']].mean()
print(mean)
```
