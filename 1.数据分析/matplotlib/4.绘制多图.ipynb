# 绘制多图
- subplot() 需要指定位置
- subplots() 调用生成对象的 ax

### subplot()
- subplot(nrows, ncols, index, **kwargs)
- subplot(pos, **kwargs)
- subplot(**kwargs)
- subplot(ax)

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
xpoints = np.array([0, 6])
ypoints = np.array([0, 100])

plt.subplot(1, 2, 1)
plt.plot(xpoints,ypoints)
plt.title("plot 1")

#plot 2:
x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("plot 2")

plt.suptitle("RUNOOB subplot Test")                 # subtitle 标题 , 居中
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["Line"])
plt.grid()

plt.show()
```
