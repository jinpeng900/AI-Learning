# Markdown 绘图标记


```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([1 , 2 , 4 , 5 , 9 , 10 , 1 , 3 , 1  , 6 , 7])

plt.plot(ypoints , marker = 'o' , color = 'r' )
plt.show()
```
