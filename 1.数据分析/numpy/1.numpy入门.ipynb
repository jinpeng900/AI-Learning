### NumPy 的数组类被称为 ndarray , 通常被称为数组
- ##### ndarray对象属性有:  
    - 1. ndarray.ndim : 数组的轴(维度)数
    - 2. ndarray.shape : 数组的维度
    - 3. ndarray.size : 数组元素的总个数
    - 4. ndarray.dtype : 数组元素的类型
    - 5. ndarray.itemsize : 数组中每个元素的字节大小
    - 6. ndarray.nbytes : 数组占用的字节数
##### 数组的维度: 这是一个指示数组在每个维度上大小的整数元组.
- 例如: 一个矩阵的维度为 (m,n) , 表示它是一个 m 行 n 列的矩阵

```python
import numpy as np
a = np.arange(15).reshape(3,5)               # 生成一个从 0 到 14 的数组 , 并将其转换为一个 3 行 5 列的二维数组
print("数组的维度:" , a.shape)                  # 输出数组 a 的维度
print("数组的元素类型:" , a.dtype)                # 输出数组 a 的元素类型
print("数组的每个元素字节大小:" , a.itemsize)                # 输出数组 a 中每个元素的字节大小
print("数组占用的字节数:" , a.nbytes)                # 输出数组 a 占用的字节数
print("数组的元素总个数:" , a.size)                # 输出数组 a 的元素总个数
print("数组的轴(维度)数:" , a.ndim)                # 输出数组 a 的轴(维度)数
print("数组a:" , a)
```

```python
# 创建一个数组
b = np.array([1,2,3,4,5])
print("数组b:" , b)
print("数组b类型:" , type(b))
```

```python
# 创建随机矩阵
# 生成指定维度大小(3行4列) 的随机多维浮点型数据(二维) , rand 固定区间为 0.0 ~ 1.0
arr = np.random.rand(3 , 4)
print(arr)
print(type(arr))
print("-----------------------------------------------------------")

# 生成指定维度大小(3行4列) 的随机多维整型数据(二维) , randint() 可指定区间(-1 , 5)
arr = np.random.randint(-1 , 5 , size = (3 ,4))
print(arr)
print(type(arr))
print("-----------------------------------------------------------")

# 生成指定维度大小(3行4列) 的随机多维浮点型数据(二维) , uniform() 可指定区间(-1.0 , 5.0) 产生 -1 到 5 之间的均匀分布的随机数
arr = np.random.uniform(-1.0 , 5.0 , size = (3 ,4))
print(arr)
print(type(arr))
print("-----------------------------------------------------------")
```

### 创建NumPy的ndarray对象
- array() 函数 , 把 Python 列表 -> ndarray对象

```python
# 1. 创建python列表对象
my_list = [[1 , 2 , 3],
           [4 , 5 , 6],
           [7 , 8 , 9]]
arr = np.array(my_list)
print(arr)
```

### arange() 函数 , 创建 ndarray对象
- arange(起始 , 结束 , 步长 , dtype = 类型) , 类似于 range() 函数 , 但返回的是 ndarray 对象

```python
arr = np.arange(0 , 15 , 3 , dtype=int)
print(arr)
print(f"元素的类型为{arr.dtype}")
print("-------------------------------")
arr = np.arange(15).reshape(3 , 5)
print(arr)
```

### numpy 的类型转换 astype()

```python
arr1 = np.arange(0 , 10 , 2 , dtype = np.int64)
print(arr1)
arr2 = arr1.astype(np.float32)
print(arr2)
print(arr2.dtype)
```
