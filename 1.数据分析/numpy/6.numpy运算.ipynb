# Numpy 运算

##### 1 基本运算
- 数组的算数运算是按照元素的. 新的数组被创建并且被结果填充

```python
a = np.array([20 , 30 , 40 , 50])
b = np.arange(4)
c = a - b
d = a + b
print("数组a:" , a)
print("数组b:" , b)
print("数组c:" , c)
print("数组d:" , d)
```

##### 2 矩阵乘法
- 矩阵乘法可以使用 dot() 函数或 @ 运算符进行计算

```python
x = np.array([[1 , 2 , 3] , [4 , 5 , 6]])
y = np.array([[6 , 23] , [-1 , 7] , [8 , 9]])
print(x.dot(y))
print("----------------------")
print(x @ y)
print("----------------------")
print(np.dot(x , y))
```
