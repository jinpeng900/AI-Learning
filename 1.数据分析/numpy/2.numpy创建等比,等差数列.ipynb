### 等比数列
- np.logspace(start , stop , num , base , dtype)
    - start : 序列的起始值 , 默认值为 0
    - stop : 序列的结束值
    - num : 要生成的样本数量 , 默认值为 50
    - base : 对数空间的底数 , 默认值为 10
    - dtype : 输出数组的数据类型 , 默认值为 None

```python
# 导包
import numpy as np

a = np.logspace(0 , 0 , 10)
a
```

```python
b = np.logspace(0 , 2 , 3 , base=3 , dtype = int)
b
```

```python
c = np.logspace(0 , 9 , 10 , base=2 , dtype=int)
c
```

### 等差数列
- np.linspace(start , stop , num , endpoint , dtype)
    - start : 序列的起始值
    - stop : 序列的结束值
    - num : 要生成的样本数量 , 默认值为 50
    - endpoint : 如果为 True , 则 stop 是序列的最后一个值 , 默认值为 True
    - dtype : 输出数组的数据类型 , 默认值为 None

```python
a = np.linspace(1 , 10 , 10 , True , dtype=int)
a
```

```python
b = np.arange(0 , 50 , 2)                   # 生成 0 到 48 的偶数 , arange() 也可以用作等差数列
b
```
