# 距离度量：量化万物之间的相似性

在机器学习中，几乎所有的算法都依赖于“距离”或“相似度”的概念。如果两个样本在特征空间中距离很近，我们就认为它们很相似。

---

## 1. 闵可夫斯基距离 (Minkowski Distance)

### 定义
它是欧氏距离和曼哈顿距离的推广形式：
$$D(x, y) = \left( \sum_{i=1}^n |x_i - y_i|^p \right)^{1/p}$$
- 当 $p=1$ 时，为**曼哈顿距离**。
- 当 $p=2$ 时，为**欧氏距离**。
- 当 $p \to \infty$ 时，为**切比雪夫距离**。

### 生动比喻
$p$ 就像是一个**“惩罚权重”**。$p$ 越大，距离计算就越关注那些差异最大的维度。

### 手写实现：通用闵氏距离函数

```python
import numpy as np
import matplotlib.pyplot as plt

def manual_minkowski(x, y, p):
    """手写闵氏距离"""
    return np.sum(np.abs(x - y)**p)**(1/p)

p1 = np.array([1, 2])
p2 = np.array([4, 6])

print(f"p=1 (Manhattan): {manual_minkowski(p1, p2, 1)}")
print(f"p=2 (Euclidean): {manual_minkowski(p1, p2, 2)}")
```

## 2. 余弦相似度 (Cosine Similarity)

### 定义
衡量两个向量方向的夹角，而不是数值大小：
$$\cos(\theta) = \frac{A \cdot B}{\|A\| \|B\|}$$

### 生动比喻
它是**志同道合**的程度。两个人可能能力大小不同（向量模长不同），但如果他们的目标一致（方向一致），余弦相似度就很高。常用于文本分类和推荐系统。

### 小数据案例：文本向量对比

```python
from sklearn.metrics.pairwise import cosine_similarity

def manual_cosine(a, b):
    """手写余弦相似度"""
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot_product / (norm_a * norm_b)

v1 = np.array([1, 1, 0]) # 喜欢看电影和看书
v2 = np.array([2, 2, 0]) # 非常喜欢看电影和看书（频率更高）
v3 = np.array([0, 0, 1]) # 喜欢运动

print(f"v1 与 v2 的余弦相似度: {manual_cosine(v1, v2):.2f} (完全方向一致)")
print(f"v1 与 v2 的欧氏距离: {np.linalg.norm(v1 - v2):.2f} (有数值差异)")
print(f"v1 与 v3 的余弦相似度: {manual_cosine(v1, v3):.2f} (完全不相关)")
```

## 3. 应用案例：手写一个简单的 KNN 逻辑

```python
def simple_knn(train_X, train_y, test_x, k=3):
    """
    基于欧氏距离的手写 KNN
    """
    distances = [np.linalg.norm(x - test_x) for x in train_X]
    # 获取最近的 k 个样本的索引
    k_idx = np.argsort(distances)[:k]
    # 多数表决
    k_labels = [train_y[i] for i in k_idx]
    return max(set(k_labels), key=k_labels.count)

# 小规模模拟数据
train_X = np.array([[1, 2], [2, 3], [3, 3], [6, 5], [7, 7], [8, 6]])
train_y = ['A', 'A', 'A', 'B', 'B', 'B']
test_point = np.array([3, 4])

prediction = simple_knn(train_X, train_y, test_point)
print(f"点 [3, 4] 的预测类别为: {prediction}")

# 可视化数据点
plt.scatter(train_X[:3, 0], train_X[:3, 1], c='red', label='Class A')
plt.scatter(train_X[3:, 0], train_X[3:, 1], c='blue', label='Class B')
plt.scatter(3, 4, c='green', marker='x', s=100, label='Test Point')
plt.legend()
plt.title("Handwritten KNN logic visualization")
plt.show()
```
