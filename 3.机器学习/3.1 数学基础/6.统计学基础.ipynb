# 统计学基础：从数据描述到推断决策

统计学是机器学习的母体。它帮助我们描述数据分布、量化变量间的关系，并从样本推断总体。

---

## 1. 描述性统计 (Descriptive Statistics)

### 定义
- **均值 (Mean)**：观测值的平均。 $\mu = \frac{1}{n} \sum x_i$
- **方差 (Variance)**：衡量数据的离散程度。 $\sigma^2 = \frac{1}{n} \sum (x_i - \mu)^2$
- **协方差 (Covariance)**：衡量两个变量的变化趋势是否一致。 $Cov(X, Y) = \frac{1}{n} \sum (x_i - \mu_x)(y_i - \mu_y)$
- **相关系数 (Correlation)**：标准化的协方差，范围 [-1, 1]。 $\rho = \frac{Cov(X, Y)}{\sigma_x \sigma_y}$

### 生动比喻
- **均值**是队伍的平均身高。
- **方差**是队员们高矮胖瘦的参差程度。
- **相关系数**就像是双人舞：如果两人动作完全同步，系数为 1；如果反向同步，为 -1；如果各跳各的，为 0。

### 手写实现：统计指标计算

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def manual_stats(data_x, data_y):
    """
    手写计算均值、方差、协方差和相关系数
    """
    n = len(data_x)
    mean_x, mean_y = np.mean(data_x), np.mean(data_y)
    
    var_x = np.sum((data_x - mean_x)**2) / n
    var_y = np.sum((data_y - mean_y)**2) / n
    
    cov_xy = np.sum((data_x - mean_x) * (data_y - mean_y)) / n
    corr_xy = cov_xy / (np.sqrt(var_x) * np.sqrt(var_y))
    
    return mean_x, var_x, cov_xy, corr_xy

# 小数据案例：学习时间与考试分数
X_study = np.array([2, 4, 6, 8, 10])
Y_score = np.array([40, 55, 65, 80, 95])

m, v, c, r = manual_stats(X_study, Y_score)
print(f"学习时间均值: {m}, 分数方差: {v}")
print(f"协方差: {c}, 相关系数: {r:.4f}")
```

## 2. 参数估计 (Parameter Estimation)

### 定义
- **极大似然估计 (MLE)**：选择能使观测数据出现概率最大的参数 $\theta$。
- **最大后验估计 (MAP)**：在 MLE 的基础上加入先验知识（正则化的数学来源）。

### 手写实现：正态分布的均值 MLE 估计

```python
def mle_normal_mu(samples):
    """正态分布均值的 MLE 就是样本均值"""
    return np.sum(samples) / len(samples)

true_mu = 50
samples = np.random.normal(true_mu, 5, 100)
est_mu = mle_normal_mu(samples)
print(f"真实均值: {true_mu}, MLE 估计均值: {est_mu:.4f}")
```

## 3. 假设检验 (Hypothesis Testing)

### 核心概念
- **零假设 ($H_0$)**：通常指“没有差异”或“没有影响”。
- **P值 (P-value)**：在 $H_0$ 成立的情况下，观察到当前数据（或更极端数据）的概率。如果 $P < 0.05$，我们拒绝 $H_0$。

### 生动比喻
假设检验就像是**法庭审判**。法官首先假设被告是无辜的（$H_0$）。如果控方提供的证据在被告无辜的前提下几乎不可能发生（$P$ 值极小），那么法官就会拒绝无辜假设，判定有罪。

### 小数据案例：新药是否有效？
对比两组病人的恢复时间。

```python
# A组：使用常规药物的恢复天数
group_a = [10, 12, 9, 11, 13, 10, 12]
# B组：使用新药的恢复天数
group_b = [8, 7, 9, 8, 10, 7, 9]

t_stat, p_val = stats.ttest_ind(group_a, group_b)
print(f"T统计量: {t_stat:.4f}, P值: {p_val:.4f}")

if p_val < 0.05:
    print("结论：在 5% 的显著性水平下，拒绝原假设，新药有效！")
else:
    print("结论：无法拒绝原假设，新药效果不显著。")
```
