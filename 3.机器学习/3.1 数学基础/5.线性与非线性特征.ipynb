# 线性特征与非线性特征：从简单映射到高维空间

在机器学习中，特征的表达能力决定了模型的上限。线性模型只能处理线性可分的数据，而通过非线性特征变换，我们可以解决更复杂的问题。

---

## 1. 线性特征 (Linear Features)

### 定义
如果一个特征是输入变量的**线性组合**，即满足叠加原理和齐次性，则称其为线性特征：
$$y = w_1 x_1 + w_2 x_2 + ... + w_n x_n + b$$

### 生动比喻
线性特征就像是**调制鸡尾酒**。你按比例加入伏特加、果汁和糖浆，最终的味道是这些成分简单叠加的结果。你多加一倍果汁，果汁的味道就增强一倍。

### 手写实现：线性特征组合

```python
import numpy as np
import matplotlib.pyplot as plt

def manual_linear_combination(X, weights, bias):
    """
    手写线性组合：y = XW + b
    """
    if X.shape[1] != len(weights):
        raise ValueError("特征数与权重数不匹配")
    
    # 使用矩阵乘法实现线性组合
    return np.dot(X, weights) + bias

# 小数据案例：房屋价格预测（线性模型）
# 特征：[面积(平米), 房间数]
X_house = np.array([[80, 2], [120, 3], [60, 1], [150, 4]])
weights = np.array([5.0, 10.0]) # 每平米5k，每个房间10k
bias = 50.0                    # 基础装修费 50k

prices = manual_linear_combination(X_house, weights, bias)
print("预测房屋价格 (k):", prices)
```

## 2. 非线性特征 (Non-linear Features)

### 定义
当特征与输入变量之间不是比例关系，或者包含了自变量的高次项、交互项或非线性函数（如 $\sin, \log, \exp$）时，称为非线性特征：
$$y = f(x) \quad \text{where } f \text{ is non-linear (e.g., } x^2, \sin(x), x_1x_2)]$$

### 生动比喻
非线性特征就像是**化学反应**。氢气和氧气结合生成的不是混合气体，而是性质完全不同的水。或者像**折纸**：通过折叠（非线性变换），一张平面的纸可以变成三维的纸飞机，从而在三维空间中实现“跨越”。

### 关键技术：多项式特征变换 (Polynomial Features)
将低维线性不可分的数据映射到高维空间，使其变为线性可分。

### 手写实现：二阶多项式变换

```python
def manual_poly_features(X):
    """
    手写二阶多项式变换：[x1, x2] -> [1, x1, x2, x1^2, x1*x2, x2^2]
    """
    n_samples, n_features = X.shape
    # 结果包含：常数项(1), 原始项(n), 平方项(n), 交互项(nC2)
    res = [np.ones(n_samples)] # 常数项
    
    # 原始项
    for i in range(n_features):
        res.append(X[:, i])
    
    # 平方项与交互项
    for i in range(n_features):
        for j in range(i, n_features):
            res.append(X[:, i] * X[:, j])
            
    return np.array(res).T

X_simple = np.array([[2, 3]])
X_poly = manual_poly_features(X_simple)
print("原始特征:", X_simple)
print("多项式变换后的特征:", X_poly)
print("顺序: [1, x1, x2, x1^2, x1*x2, x2^2]")
```

## 3. 可视化：线性不可分到线性可分

### 小数据案例：环形数据集
在 2D 平面上，内圆和外圆是线性不可分的。但如果我们引入 $z = x^2 + y^2$，它们在 3D 空间中就可以被一个平面切开。

```python
from sklearn.datasets import make_circles

# 生成环形数据
X, y = make_circles(n_samples=200, factor=0.5, noise=0.05, random_state=42)

fig = plt.figure(figsize=(12, 5))

# 2D 视图
ax1 = fig.add_subplot(121)
ax1.scatter(X[y==0, 0], X[y==0, 1], color='red', label='Class 0')
ax1.scatter(X[y==1, 0], X[y==1, 1], color='blue', label='Class 1')
ax1.set_title("2D View: Linearly Inseparable")
ax1.legend()

# 3D 变换视图：增加 z = x^2 + y^2
ax2 = fig.add_subplot(122, projection='3d')
Z = X[:, 0]**2 + X[:, 1]**2
ax2.scatter(X[y==0, 0], X[y==0, 1], Z[y==0], color='red')
ax2.scatter(X[y==1, 0], X[y==1, 1], Z[y==1], color='blue')
ax2.set_title("3D Transformation: Linearly Separable")
plt.show()
```

## 4. 库函数对比：Sklearn PolynomialFeatures

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)
X_poly_sklearn = poly.fit_transform(X_simple)
print("Sklearn 多项式变换结果:\n", X_poly_sklearn)
print("结果是否一致:", np.allclose(X_poly, X_poly_sklearn))
```
